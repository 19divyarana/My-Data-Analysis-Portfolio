```python
#Problem 1 – String slicing & methods (very easy)
text = "Hyderabad is a great city to learn data analysis"
# 1. Print only "great city"
print(text[15:25])
# 2. Print the string in uppercase
print(text.upper())
# 3. Count how many times the letter 'a' appears
print(text.count("a"))
# 4. Replace "great" with "awesome"
print(text.replace("great","awesome"))
# 5. Check if the string ends with "analysis"
print(text.endswith("analysis"))
```

```python
#Problem 2 – List basics + indexing
cities = ["Hyderabad", "Bangalore", "Chennai", "Pune", "Mumbai", "Delhi"]

# 1. Print the 3rd city
print(cities[2])
# 2. Print the last two cities using negative indexing
print(cities[-2:])
# 3. Add "Kolkata" to the end of the list
cities.append("Kolkata")
print(cities)
# 4. Insert "Ahmedabad" at position 2 (second place)
cities.insert(2,"Ahmedabad")
print(cities)
# 5. Remove "Pune" from the list
cities.remove("Pune")
print(cities)
# 6. Print the length of the list after changes
print(len(cities))
```

```python
#Problem 3 – List + loop (simple iteration)

numbers = [4, 17, 8, 23, 9, 12, 5, 14]

# Print only the even numbers from the list
for num in numbers:
    if num % 2 == 0:
        print(num)
# (use a for loop and if condition)
```

```python
#Problem 4 – Tuple (immutability awareness)
student = ("Divya", 21, "CSE-AIML", "Hyderabad", 8.5)

# 1. Print the entire tuple
print(student)
# 2. Print only the name and CGPA
print("2. Name and CGPA:", student[0], "with CGPA", student[4])
# 3. Try to change the CGPA to 9.0 → what happens? (write comment)
#we cannot chnage -so it will give an error
# 4. Convert the tuple to a list, change CGPA to 9.0, then convert back to tuple
stu = list(student)
stu[4]=9.0
student_list= tuple(stu)
print(student_list)
```
