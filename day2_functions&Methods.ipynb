```python
# Day 2: Functions + Methods practice for data analyst journey
# Problem 1 – Simple function with string methods
def clean_city_name(city):
    """
    Take a city name (possibly messy), return it cleaned:
    - Remove extra spaces
    - Capitalize first letter only (title case)
    """
    cleaned = city.strip().title()          # strip removes leading/trailing spaces, title makes "Hyderabad"
    return cleaned

print("Problem 1:", clean_city_name("  bangaLORE  "))      # → Bangalore
print("Problem 1:", clean_city_name("  hyderabad   "))     # → Hyderabad
print()
```

```python
# Problem 2 – Function that counts vowels in a string
def count_vowels(text):
    """Return number of vowels (a,e,i,o,u) in the text, case insensitive"""
    text_lower = text.lower()
    count = 0
    for vowel in 'aeiou':
        count += text_lower.count(vowel)
    return count

print("Problem 2:", count_vowels("Hyderabad Data Analysis"))   # → 9
print("Problem 2:", count_vowels("SQL and Python"))            # → 3
print()
```

```python
def add_unique_skill(skills_list, new_skill):
    """
    Add a skill to the list ONLY if it's not already there.
    Return the updated list.
    """
    if new_skill not in skills_list:
        skills_list.append(new_skill)
    return skills_list

my_skills = ["Python", "SQL"]
print("Problem 3 - Before:", my_skills)
add_unique_skill(my_skills, "Python")     # no change
print("After adding Python again:", my_skills)
add_unique_skill(my_skills, "Pandas")
print("After adding Pandas:", my_skills)
print()
```

```python
# Problem 4 – String split + list methods + set for unique
def extract_domains(emails_string):
    """
    Input: string like "divya@gmail.com, raj@outlook.com, priya@yahoo.in"
    Return a list of unique domains only
    """
    emails = emails_string.split(", ")                  # split by comma + space
    domains = set()                                     # use set to keep unique
    for email in emails:
        if '@' in email:
            domain = email.split('@')[1]                # take part after @
            domains.add(domain)
    return sorted(list(domains))                        # return sorted list

test_emails = "divya@gmail.com, raj@outlook.com, priya@yahoo.in, divya@gmail.com"
print("Problem 4:", extract_domains(test_emails))
# → ['gmail.com', 'outlook.com', 'yahoo.in']
print()
```

```python
# Problem 5 – Dictionary methods + function
def merge_profiles(profile1, profile2):
    """
    Merge two dictionaries. If keys overlap, keep value from profile2.
    """
    merged = profile1.copy()            # don't modify original
    merged.update(profile2)             # update adds/changes keys from profile2
    return merged

p1 = {"name": "Divya", "city": "Guntur"}
p2 = {"city": "Hyderabad", "skills": ["Python"]}
print("Problem 5:", merge_profiles(p1, p2))
# → {'name': 'Divya', 'city': 'Hyderabad', 'skills': ['Python']}
print()
```

```python
# Problem 6 – Set operations in a function
def common_tools(tools_a, tools_b):
    """
    Return a sorted list of tools that are in BOTH sets.
    """
    common = tools_a.intersection(tools_b)
    return sorted(list(common))

print("Problem 6:", common_tools(
    {"Excel", "SQL", "Python"},
    {"Python", "Power BI", "SQL"}
))  # → ['Python', 'SQL']
print()
```
